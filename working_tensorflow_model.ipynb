```python
import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


def custom_accuracy(y_true, y_pred):
  y_pred = tf.cast(y_pred, tf.int32)
  y_true = tf.cast(y_true, tf.int32)
  correct_predictions = tf.equal(y_true, y_pred)
  accuracy = tf.reduce_mean(tf.cast(correct_predictions, tf.float32))
  return accuracy


# Load the text based dataset
df = pd.DataFrame(data)

# Convert the conditions to numerical values
encoder = LabelEncoder()
df['condition'] = encoder.fit_transform(df['condition'])

# Convert the symptoms to numerical values using a tokenizer
symptom_texts = df['symptoms'].tolist()
tokenizer = Tokenizer()
tokenizer.fit_on_texts(symptom_texts)
sequences = tokenizer.texts_to_sequences(symptom_texts)

# Pad the sequences to the same length
padded_sequences = pad_sequences(sequences)

# Split the data into features and labels
X = padded_sequences
y = df['condition'].values

# One-hot encode the labels

y = tf.keras.utils.to_categorical(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

# Build a neural network model using Tensorflow
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(input_dim= len(tokenizer.word_index)+ 1, output_dim=32, input_length=X.shape[1]))
model.add(tf.keras.layers.LSTM(128, return_sequences=True))
model.add(tf.keras.layers.LSTM(64))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(y.shape[1], activation='softmax'))


# Compile the model loss = 

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=[custom_accuracy])
x_val = 'symptoms'
y_val = 'condition'

# Train the model
#history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data= (X_test, y_test))
model.fit(X_train, y_train, epochs=50, batch_size=32,validation_data=(X_test, y_test))
# Evaluate the model on the test data

#test_loss, test_acc = model.evaluate(X_test, y_test)
#print(test_loss)
#print(test_acc)

input_shape = model.layers[0].input_shape
print(input_shape)
model.save('health_condition_model.h5')

input_shape.shape
```
